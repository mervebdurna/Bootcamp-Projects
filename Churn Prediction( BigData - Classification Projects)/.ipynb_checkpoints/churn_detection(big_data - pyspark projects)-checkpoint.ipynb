# Data Understanding
Şirketi terk edecek müşterileri tahmin edebilecek bir makine öğrenmesi modeli geliştirebilir misiniz?

Amaç bir bankanın müşterilerinin bankayı terk etme ya da terk etmeme durumunun tahmin edilmesidir.

Müşteri terkini tanımlayan olay müşterinin banka hesabını kapatmasıdır.

Veri Seti Hikayesi:

10000 gözlemden ve 12 değişkenden oluşmaktadır.
Bağımsız değişkenler müşterilere ilişkin bilgiler barındırmaktadır.
Bağımlı değişken müşteri terk durumunu ifade etmektedir.
Değişkenler:

Surname : Soy isim

CreditScore : Kredi skoru

Geography : Ülke (Germany/France/Spain)

Gender : Cinsiyet (Female/Male)

Age : Yaş

Tenure : Kaç yıllık müşteri

Balance : Bakiye

NumOfProducts : Kullanılan banka ürünü

HasCrCard : Kredi kartı durumu (0=No,1=Yes)

IsActiveMember : Aktif üyelik durumu (0=No,1=Yes)

EstimatedSalary : Tahmini maaş

Exited : Terk mi değil mi? (0=No,1=Yes)

### Outliers 

```python
cat_features = ['Geography','Gender','NumOfProducts','HasCrCard','IsActiveMember']# 'CustomerId','Surname', 
num_features = ['CreditScore','Age','Tenure','Balance','EstimatedSalary']
```

```python
df = spark_df.toPandas()
```

```python
df.head()
```

```python
from sklearn.neighbors import LocalOutlierFactor

clf=LocalOutlierFactor(n_neighbors=20, contamination=0.1)
clf.fit_predict(df[num_features])
df_scores=clf.negative_outlier_factor_
df_scores= np.sort(df_scores)
df_scores[0:20]
```

```python
sns.boxplot(df_scores);
```

```python
threshold=np.sort(df_scores)[7]
print(threshold)
df = df.loc[df_scores > threshold]
df = df.reset_index(drop=True)
```

```python
spark_df = spark.createDataFrame(df)
spark_df.show()
```

## Feature Engineering

```python
# ODEV :   df['New_IsBalanceZero'] = df['Balance'].map(lambda x: 1 if x == 0 else 0)
spark_df = spark_df.withColumn("New_TenureAgeRatio", spark_df['Tenure']/spark_df['Age'])
spark_df = spark_df.withColumn("New_BalanceEstimatedSalaryRatio", spark_df['Balance']/spark_df['EstimatedSalary'])
spark_df = spark_df.withColumn("New_BalanceCreditScoreRatio", spark_df['Balance']/spark_df['CreditScore'])
```

## Setting Label

```python
# Bagimli degiskeni belirlemek 
from pyspark.ml.feature import StringIndexer
# StringIndexer fonksiyonu string formattaki bagımlı değişkendeki en çok olan sınıfa 0 diğerine 1 verir.
stringIndexer = StringIndexer(inputCol = "Exited", outputCol = "label")
```

```python
indexed = stringIndexer.fit(spark_df).transform(spark_df)
```

```python
# Indexed datasetindeki değişken türlerine bakınca label double türünde bunun integer olması lazım.
indexed.dtypes
```

```python
spark_df = indexed.withColumn("label", indexed["label"].cast("integer"))
```

```python
spark_df.dtypes
```

```python
spark_df.show()
```

## Setting Features

```python
# bagimsiz degiskenleri belirlemek : spark modelleri bağımsız değişkenleri vektör şeklinde girdi olarak kabul ediyor.
from pyspark.ml.feature import VectorAssembler
```

```python
spark_df.dtypes
```

### Label Encoding 

```python
# geography değişkenini sayısal değerlere çevirdik yani label_encoder uyguladık 
# ve bunu geography_categoryIndex değişkenine atadık
indexer = StringIndexer(inputCol="Geography", outputCol="Geography_CategoryIndex")
indexed1 = indexer.fit(spark_df).transform(spark_df)

# Double türündeki gender_categoryIndex değişkeninin türünü integer yapar.
indexed1 = indexed1.withColumn("Geography_CategoryIndex", indexed1["Geography_CategoryIndex"].cast("integer"))
```

```python
indexed1.dtypes
```

```python
indexer = StringIndexer(inputCol="Gender", outputCol="Gender_CategoryIndex")
indexed1 = indexer.fit(indexed1).transform(indexed1)
indexed1 = indexed1.withColumn("Gender_CategoryIndex", indexed1["Gender_CategoryIndex"].cast("integer"))
```

```python
indexed1.show(5)
```

### One Hot Encoding

```python
# geography_categoryIndex değişkeninin frekansına baktık. Bu kategorik bir değişken ve frekansı 2'den büyük ...
# ... olduğu için buna one_hot_encoding uygulanmalı
indexed1.select("Geography_CategoryIndex").distinct().count()
```

```python
# geography_categoryIndex değişkenine one_hot_encoding uyguladık ve data frame i "indexed2" diye isimlendirdik.
from pyspark.ml.feature import OneHotEncoder
encoder = OneHotEncoder(inputCols=["Geography_CategoryIndex"],
                        outputCols=["Geography_CategoryIndex_vec"])

indexed2 = encoder.fit(indexed1).transform(indexed1)
```

```python
indexed2.show()
```

```python
indexed2.dtypes
```

```python
indexed2.columns
```

```python
columns=['CreditScore',
         'Age',
         'Tenure',
         'Balance',
         'NumOfProducts',
         'HasCrCard',
         'IsActiveMember',
         'EstimatedSalary',
         'New_TenureAgeRatio',
         'New_BalanceEstimatedSalaryRatio',
         'New_BalanceCreditScoreRatio',
         'Geography_CategoryIndex_vec']
```

```python
# Bağımsız değişkenleri vektörleştirecek nesne oluşturulur.
va = VectorAssembler(inputCols = columns, outputCol = "features")
```

```python
# Bagımsız değişkenlerin vektöre çevirir.
va_df = va.transform(indexed2)
```

```python
va_df.show()
```

```python
display(va_df)
```

```python
# Dönüştürülen bağımlı ve bağımsız değişkenler bir araya getirilir.
final_df = va_df.select(["features","label"])
```

```python
final_df.show()
```

### Feature Scaling

```python
from pyspark.ml.feature import StandardScaler
scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures",
                        withStd=True, withMean=False)
final_scaled_df = scaler.fit(final_df).transform(final_df)
```

## Modelling - Pyspark

```python
splits = final_scaled_df.randomSplit([0.80, 0.20],seed=12345)
train_df = splits[0]
test_df = splits[1]
```

### GBM Base Model - PySpark

```python
#base model GBM
from pyspark.ml.classification import GBTClassifier
gbm = GBTClassifier(maxIter = 10, featuresCol = "features", labelCol = "label")
gbm_model = gbm.fit(train_df)
y_pred = gbm_model.transform(test_df)
```

```python
ac = y_pred.select("label","prediction")
```

```python
ac.show(3)
```

```python
ac.filter(ac.label == ac.prediction).count()/ac.count()
```

### GBM Moddel Tuning - PySpark

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

evaluator = BinaryClassificationEvaluator()

paramGrid = (ParamGridBuilder()
             .addGrid(gbm.maxDepth, [2, 4, 6])
             .addGrid(gbm.maxBins, [20, 30])
             .addGrid(gbm.maxIter, [10, 20])
             .build())

cv = CrossValidator(estimator = gbm, estimatorParamMaps = paramGrid, evaluator = evaluator, numFolds = 10)
```

```python
cv_model = cv.fit(train_df)
```

```python
y_pred = cv_model.transform(test_df)
```

```python
ac = y_pred.select("label","prediction")
```

```python
ac.filter(ac.label == ac.prediction).count() / ac.count()
```

### Random Forest Base Model- PySpark 

```python
from pyspark.ml.classification import RandomForestClassifier

rf = RandomForestClassifier(featuresCol = "features", labelCol = "label")
rf_model =rf.fit(train_df)
y_pred = rf_model.transform(test_df)
```

```python
ac = y_pred.select("label","prediction")
```

```python
ac.show(3)
```

```python
ac.filter(ac.label == ac.prediction).count()/ac.count()
```

## Modelling- Python

```python
df = indexed2.toPandas()
```

```python
columns=['CreditScore',
         'Age',
         'Tenure',
         'Balance',
         'NumOfProducts',
         'HasCrCard',
         'IsActiveMember',
         'EstimatedSalary',
         'New_TenureAgeRatio',
         'New_BalanceEstimatedSalaryRatio',
         'New_BalanceCreditScoreRatio',
         'Geography_CategoryIndex'] # Geography_CategoryIndex_vek vektor tipinde oldugu icin burada kullanamadik. Onun yerine Geography_CategoryIndex kullandik yani ohe yerine le uyguladik


from sklearn.model_selection import train_test_split
y = df["Exited"]
X = df[columns]
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.20,random_state=12345)
```

### GBM Base Model - Python

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import accuracy_score

gbm_model = GradientBoostingClassifier().fit(X_train,y_train)
y_pred = gbm_model.predict(X_test)
```

```python
acc = np.mean(cross_val_score(gbm_model, X_train, y_train, cv = 10, scoring = "accuracy"))
acc
```

### Random Forests Base Model - Python

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier().fit(X_train,y_train)
y_pred = rf_model.predict(X_test)
```

```python
acc = accuracy_score(y_pred,y_test)
acc
```
