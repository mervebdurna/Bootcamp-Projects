```python
#Importing Librarires
import pandas as pd
import numpy as np

from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

pd.set_option('max_columns', 220)
pd.set_option('max_rows', 999)
```

```python
# Loading Dataset
df = pd.read_csv('grocery_store.csv',names=['products'],header=None)
df.columns
```

```python
df.head()
```

```python
df.shape
```

```python
df.values
```

```python
baskets = list(df['products'].apply(lambda x : x.split(',')))
baskets
```

```python
te = TransactionEncoder()
te_baskets = te.fit(baskets).transform(baskets)
baskets_df = pd.DataFrame(te_baskets,columns=te.columns_)
baskets_df
```

```python
#use_colnames = True, verbose = 1
freq_items = apriori(baskets_df, min_support=0.04, use_colnames=True)
freq_items.sort_values(by="support",ascending=False)
```

```python
rules = association_rules(freq_items, metric = "confidence", min_threshold = 0.6)
rules
```

```python
## Yorumlar
```

```python
rules.sort_values(by="support",ascending=False)
```

```python
rules.sort_values(by="confidence",ascending=False)
```

```python
rules.sort_values(by="lift",ascending=False)
```

```python
# Enbuyuk support degerine sahip urunleri yorumlama

(support) --> CORNFLAKES ve COFFEE tm alverilerin yzde 20 sinde birlikte gozlemlenmektedir
(confidence) --> CORNFLAKES alan mterilerin yzde 66 si COFFEE de almaktadir
(lift) --> CORNFLAKES olan alverilerde COFFEE rnnn sat 1.66 kat artmaktadr
```

```python
# Enbuyuk confidence degerine sahip urunleri yorumlama
# Enbuyuk lift degerine sahip urunleri yorumlama
(MILK, COFFEE)	(TEA, CORNFLAKES) :  (support) tm alverilerin yuzde 5 inde (MILK, COFFEE) ve (TEA, CORNFLAKES) birlikte gozlemlenmektedir.
                                     (confidence) --> (MILK, COFFEE) alan mterilerin yzde 100u  (TEA, CORNFLAKES) de almaktadir
                                     (lift) --> (MILK, COFFEE) olan alverilerde (TEA, CORNFLAKES) rnnn sat 10 kat artmaktadr 
```
