# Business Problem 

Company X, which adopts the MAXIMUM bidding strategy for website advertisements, wants to statistically test whether the AVERAGE bidding strategy introduced recently will bring more conversion rates and purchases.

For this reason, the target audience is divided into two groups of equal size (control - test). The maximum bidding strategy is applied to the control group and the average bidding strategy is applied to the test group for one month. It is tested whether there is a statistically significant difference between the two groups with the A / B test on the conversion rates and purchases obtained at the end of 1 month.

# Applying Hypothesis Testing

# Parametric Two  Samples Independent T Test (AB Test)

H0 : There is **no significant** difference between the mean of Purchase of the two groups(A-B : AVG Bidding - MAX - Bidding). (mA = mB)

H1 : There is **significant** difference between the mean of Purchase of the two groups(A-B : AVG Bidding - MAX - Bidding). (mA != mB)

## Assumptions : 

* **Normality :** The data for each group should be approximately normally distributed. 
* **Homogeneity of variances :**  The variance of the outcome variable should be equal in each group. Recall that, the Welch t-test does not make this assumptions.

```python
A = df_AB[df_AB["Group"] =="A"]['Purchase']
B = df_AB[df_AB["Group"] =="B"]['Purchase']
```

```python
def checkReject(p_value, alpha = 0.05, one_tailed = True):
    if one_tailed == False :
        alpha = alpha/2
    
    if p_value < alpha :
        print('Reject to H0 Hypothesis.')
    
    elif p_value > alpha : 
        print('Fail to reject H0 Hypothesis.')
```

## Normality

**Shapiro-Wilk Test :** Test of normality.  
    
    H0 : The sample is drawn from a normally distributed population.
    H1 : The sample is drawn from a population that is NOT normally distributed. 
    
    alpha = 0.05 (significance level) --> one-tailed test
    alpha/ 2 = 0.025 --> two tailed test.
    
    p < alpha : Reject the H0.(**significant**) (We have enough evidence to say that the population is not normally distributed.)
    
    p > alpha : Fail to reject H0.(**non-significant**)(We don'thave enough evidence to say that the population is not normally distributed.)
    
NOTE: Hypotheses testing; more about evidence, it doens't more about True or False
    
NOTE : To meet the normality assumption we need to fail to reject the null hypothesis. (p > 0.05)

```python
from scipy.stats import shapiro
```

```python
p_value = shapiro(A)[1]
p_value
```

```python
checkReject(p_value)
```

For A : 

We fail to reject H0 hypothesis. 

We fail to reject that "The sample is drawn from a normally distributed population".

**The normality assumption for A group has been met.**



```python
p_value = shapiro(B)[1]
p_value
```

```python
checkReject(p_value)
```

For B :

We reject H0 hypothesis.

We reject that "The sample is drawn from a normally distributed population".

**The normality assumption for B group has been violated.**

## Homogeneity of variances
**Levene Test:** Test wheter the variances of two samples are aproximately equal. (F-TEST likE ANOVA)

H0 : There is no difference between the variance of the first group and variance of the second group. 

H1 : The sample is drawn from a population that is NOT normally distributed. 

alpha = 0.05 (significance level) --> one-tailed test
alpha/ 2 = 0.025 --> two tailed test.

p < alpha : Reject the H0.(**significant**)

p > alpha : Fail to reject H0.(**non-significant**)


NOTE : We want to Levene test to be **not-significan** because we dont want the variances to be diffrent.   

NOTE : To satisfy the homogeneity to variance assumption we need to fail to reject the null hypothesis. (p > 0.05)



```python
from scipy.stats import levene
p_value = levene(A,B)[1]
p_value
```

```python
checkReject(p_value)
```

**We could not statistically reject that the variances were homogeneous. The homogeneity of variance assumption has been met.**

That's why we will aplly non-parametric hypothesis test for two sample. 


# Non-parametric Independent Two-Sample T Test (Mann Wihtney U)

Since, normality assumption was not met for the group A, Mann Wihtney U test was applied.

H0 : There is no significant difference between the mean of Purchase of the two groups(A-B : AVG Bidding - MAX - Bidding). (mA = mB)

H1 : There is significant difference between the mean of Purchase of the two groups(A-B : AVG Bidding - MAX - Bidding). (mA != mB)

```python
from scipy.stats import mannwhitneyu
```

```python
p_value = mannwhitneyu(A,B)[1]
p_value 
```

```python
checkReject(p_value)
```

**We failed to reject that there is no significant difference between the Purchase mean of two groups( Average Bidding and Max Bidding).**

# Conclusion

* It was observed that the assumption of normality of the **"Two Samples T Test"**  was not meet for the test group as a result of the shapiro test.

* Since the assumptions were not provided, the non-parametric **"Mann Wihtney U Test"** was applied.

* The H0 hypothesis failed to reject statistically because the p-value(0.17) obtained as a result of the Mann Wihtney U Test was greater than 0.05. 

* Failed to reject that There is **no significant** difference between the mean of Purchase of the two groups.

* **There is no need to take actions in favor of AVERAGE bidding strategy.**
